```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
m = 1000 # number of generation
N = 1000 # generation size

def generate_traits(correlation = 0):
    means = [0,0]
    cov = [[1,correlation],
           [correlation,1]]
    return np.random.multivariate_normal(means, cov, size=N).T

A,B = generate_traits(0.8)
X = np.zeros((m,N))
print(f'A.shape = {A.shape}\t\tB.shape = {B.shape}\t\tX.shape = {X.shape}')
```

```python
b,J = 1,0.5
A_hat = 1
alpha = 0.8
alpha_tag = alpha/(1- alpha)

print(f'b={b}\tJ={J}\tA_hat={A_hat}\talpha={alpha:.2f}\talpha_tag={alpha_tag:.2f}')
```

```python
def bias(indicator_traits):
    return b * np.exp(-(((A_hat - indicator_traits) ** 2) / (2 * J)))

def get_generation_data(gen,A_,B_):
    return pd.DataFrame([[gen,'Indicator',A_.mean(),A_.std(),A_.var()],
                         [gen,'Indirect',B_.mean(),B_.std(),B_.var()]],
                        columns = df_columns)
    
df_columns = ['Generation','Type','Mean','Std','Variance']
df = pd.DataFrame(columns=df_columns)


# fully iterative dependant model. No approximations are made
copiers = np.random.rand(X.shape[0],X.shape[1])

for i in range(m):
    biased_A = bias(A)*alpha_tag
    
    new_A,new_B = [],[]
    for j in range(N):
        prestige = biased_A + X[i,:]
        scores = prestige/prestige.sum()
        chosen_role_model = np.searchsorted(scores.cumsum(),copiers[i][j])
        X[i,chosen_role_model] += 1
        new_A.append(A[chosen_role_model])
        new_B.append(B[chosen_role_model])

    A = np.array(new_A)
    B = np.array(new_B)
    generation_data = get_generation_data(i+1,A,B)
    df = df.append(generation_data, ignore_index=True)
      
df.tail()      
```
