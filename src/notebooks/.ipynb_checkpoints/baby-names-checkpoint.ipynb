```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from glob import glob
from scipy.stats import entropy
```

```python
files = glob('us-data/*.txt')
years = range(1880, 2019)
df = pd.DataFrame()
for y in years:
    df_ = pd.read_csv('us-data/yob{}.txt'.format(y), header=None,  names=['Name', 'Sex', 'Count'])
    df_['Year'] = y
    df = pd.concat((df, df_))
```

```python
df.head()
```

```python
df['Frequency'] = df.groupby('Year')['Count'].transform(lambda x: x / x.sum())
```

```python
def diversity(x):
    return np.exp(entropy(x))

def richness(x):
    x = np.array(x)
    return (x>0).sum()
```

```python
df.groupby('Year')['Frequency'].aggregate(richness).plot.line(label='richness')
df.groupby('Year')['Frequency'].aggregate(diversity).plot.line(label='true diversity')
plt.ylabel('Entropy')
plt.legend();
```

```python
def plot_name(name, sex, logscale=False, ax=None):
    if ax is None: fig, ax = plt.subplots()
    idx = (df['Name'] == name) & (df['Sex'] == sex)
    year = df.loc[idx, 'Year']
    freq = df.loc[idx, 'Frequency']
    ax.plot(year, freq, label="{} ({})".format(name, sex))
    if logscale: ax.set_yscale('log')
    return ax
```

## Single Name Model

```python
single_name_df = df[(df['Name'] == 'Paul')&(df['Sex'] == 'M')]
P = single_name_df['Frequency'].values
Y = single_name_df['Count'].values
N = df[df['Sex'] == 'M'].groupby('Year')['Count'].sum().values
t = np.arange(len(P))

diffs = np.abs(np.diff(P)/np.diff(t))
max_diffs = np.argpartition(diffs, -3)[-3:]

plt.plot(t,P)
for diff in max_diffs:
    plt.axvline(diff,color='k',linestyle='--')
plt.show();
```

```python
def plot_model(θ,t,p0,model):
    p_hat = model(θ,t,p0)
    
    plt.plot(t,p_hat)
    plt.xlabel('Time(y)')
    plt.ylabel('Frequency')
```

## Simple Selection Model

```python
def simple_verify_params(θ):
    s = θ
    if s > -1:
        return True
    return False

def simple_step(s,p):
    w = 1 + s
    w_mean = w*p + (1-p)
    p_tag = w*p / w_mean
    return p_tag

def simple_model(θ,t,p0):
    p = np.zeros(len(t))
    p[0] = p0
    for i in t[1:]:
        p[i] = simple_step(θ,p[i-1])

    return p    
        
t = np.arange(100)
θ = (0.2)
plot_model(θ,t,0.02,simple_model);
```

## Frequency Parabola-Dependant Model

```python
from ipywidgets import interact

def verify_freq_params(θ):
    a,b,c = θ
    if c <0 and a <0 and b>0 and b < -2*a and a<c and b**2 > 4*a*c and b < -(c + a):
        return True
    return False

def frequency_step(a,b,c,p):
    s = a*(p**2) + b*p + c
    s = s if s > -1 else -0.99999
    w = 1 + s
    w_mean = w*p + (1-p)
    p_tag = w*p / w_mean
    return p_tag

def frequency_model(θ,t,p0):
    a,b,c = θ
    p = np.zeros(len(t))
    p[0] = p0
    for i in t[1:]:
        p[i] = frequency_step(a,b,c,p[i-1])

    return p    
        
t = np.arange(100)
θ = (-5,-10,4)
plot_model(θ,t,0.02,frequency_model);
    
# interact(plot_interact,
#         n = (1,200,1), p0 = (0,1,0.001), a=(-5,0,0.01),b=(-10,20,0.5),c=(-20,30,2));
```

## Fitting to the data

```python
import scipy

def rss(y,y_hat):
    return np.sum((y-y_hat)**2)

def log_likelihood(θ,Y,N,t,p0,model,verify_params): 
    n = θ[-1]
    if n <= 0 or n > N.max():
        return -np.inf
    if not verify_params(θ[:-1]):
        return -np.inf
    P_hat = model(θ[:-1],t,p0)
    return scipy.stats.binom.logpmf(Y,n,P_hat).sum()

def neg_log_likelihood(θ,Y,N,t,p0,model,verify_params):
    loss = -log_likelihood(θ,Y,N,t,p0,model,verify_params)
    return loss
```

```python
θ0 = (0.2,100000)
t = np.arange(len(P))

args = (Y,N,t,P[0],simple_model,simple_verify_params)
θ_hat_simple = scipy.optimize.fmin(
        func=neg_log_likelihood, 
        x0=θ0,
        args=args,
        maxiter=1000,
)
s1,n1 = θ_hat_simple
print(f'\n\ns = {s1}\tn = {n1}')
```

```python
θ0 = (-2,2,-0.2,100000)
t = np.arange(len(P))

args = (Y,N,t,P[0],frequency_model,verify_freq_params)
θ_hat_gauss = scipy.optimize.fmin(
        func=neg_log_likelihood, 
        x0=θ0,
        args=args,
        maxiter=10000,
)

a3,b3,c3,n3 = θ_hat_gauss
print(f'\n\na = {a3}\tb = {b3}\tc = {c3}\tn = {n3}')
```

```python
import emcee

def run_emcee(ndim,nwalkers,args,guesses):
    nsteps = 50000 // nwalkers
    nburn = nsteps // 2
    
    sampler = emcee.EnsembleSampler(
        nwalkers=nwalkers, 
        dim=ndim,
        lnpostfn=log_likelihood,
        args=args,
    )
    
    sampler.run_mcmc(
        pos0=guesses,
        N=nsteps
    )
    
    sample = sampler.chain[:, nburn:, :].reshape(-1, ndim)
    return sample
```

```python
guesses = np.empty((50,2))
guesses[:,0] = np.random.uniform(low=-0.99999,high=0,size=nwalkers)
guesses[:,1] = np.random.uniform(low=N.min(),high=N.max(),size=nwalkers)

args = (Y,N,t,P[0],simple_model,simple_verify_params)
sample = run_emcee(2,50,args,guesses)
s1,n1 = sample[:,0].mean(),sample[:,1].mean()
print(f's = {s}\tn = {n}')
```

```python
p_hat_simple = simple_model(s1,t,P[0])
p_hat_freq = frequency_model((a3,b3,c3),t,P[0])

plt.plot(t,P,label='data')
plt.plot(t,p_hat_simple,label='simple model')
plt.plot(t,p_hat_freq,linestyle='-.',label='parabola model')
plt.xlabel('Time(y)')
plt.ylabel('Frequency')
plt.legend()
plt.show();
```
