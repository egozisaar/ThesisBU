# Role-model choosing

### Model

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats

def bias_function(bias_b, bias_J, optimal_value, perceived_traits):
    return bias_b * np.exp(-(((optimal_value - perceived_traits) ** 2) / (2 * bias_J))) 

def choose_role_model(models_indicators, b=1, J=0.5, optimal_value=1,indicator_weights=0.5,eta=0.01,reps=100):
    copiers_err = np.random.normal(size=models_indicators.size, scale=eta)
    influences = np.zeros((reps,N))
    reshaped_indicators = models_indicators.reshape(1, -1)
    reshaped_copier_errs = copiers_err.reshape(-1, 1)
    perceived_indicators_matrix = reshaped_indicators + reshaped_copier_errs
    biased_indicators_matrix = bias_function(bias_b=b, bias_J=J, optimal_value=optimal_value,
                                                   perceived_traits=perceived_indicators_matrix)
    weighted_indicators_matrix = biased_indicators_matrix * indicator_weights.reshape(1, -1)
    weighted_indicators_matrix_reps = np.tile(weighted_indicators_matrix,(reps,1,1))
    random_indices = np.random.rand(reps,models_indicators.size)
    indices_row = np.arange(reps)
    for i in range(weighted_indicators_matrix_reps.shape[1]):
        weighted_influences = influences*(1-indicator_weights.reshape(1, -1))
        scores = weighted_indicators_matrix_reps[:,i,:] + weighted_influences
        prestige = scores / scores.sum(axis=1,keepdims=True)
        prestige = prestige.cumsum(axis=1)
        draws = random_indices[:,i].reshape(-1,1)
        chosen_role_model_indices = np.argmax(prestige > draws, axis=1)
        influences[indices_row,chosen_role_model_indices] += 1
    return influences

def choose_dirichlet(biased_indicators,reps=100):
    gen_size = len(biased_indicators)
    dist = np.random.dirichlet(biased_indicators,reps)
    dist =  np.tile(dist,(1,1,gen_size)).reshape(reps,gen_size,gen_size)
    dist = dist.cumsum(axis=2)
    copiers = np.random.rand(reps,gen_size).reshape(reps,gen_size,-1)
    cols = np.argmax(dist>copiers,axis=2).reshape(-1)
    rows = np.arange(reps).repeat(gen_size)
    influences = np.zeros((reps,gen_size))
    np.add.at(influences,(rows,cols),1)
    return influences

def choose_dirichlet_multinomial(biased_indicators,reps=100):
    gen_size = len(biased_indicators)
    dist = np.random.dirichlet(biased_indicators,reps)
    return np.array([np.random.multinomial(gen_size,p) for p in dist])

def dirichlet_multinomial_probs(biased_indicators):
    gen_size = len(biased_indicators)
    p = scipy.stats.dirichlet.mean(biased_indicators).reshape(-1,1)
    quntiles = np.arange(gen_size+1)
    expected_probs = scipy.stats.binom.pmf(quntiles,gen_size,p)
    return expected_probs.T


def kl_divergence(p, q):
    p[p ==0] = 1e-10
    q[q ==0] = 1e-10
    p /= p.sum()
    q /= q.sum()
    return np.sum(p * np.log(p / q))

# N = 3
# indicators = np.arange(N) + 1
# alphas = np.full(N,1)

# a= choose_role_model(indicators,indicator_weights=alphas,reps=3,optimal_value=3)
# b =  choose_dirichlet(indicators,2)
# c =  choose_dirichlet_multinomial(indicators,1)
# d = dirichlet_multinomial_probs(indicators)
# d

# binom.pmf([0,1,2],2,[[0.5],[0.6]])
```

### Parameters

```python
N = 1000
optimal_value = 0
bs = [0.01,0.1,0.5,1,5,10,100,1000]
Js = [0.01,0.1,0.5,1,5,10,100,1000]
err_scales = [0.1,0.01,0.001,0.0001]

indicators = np.random.normal(size=N)
```

The parameters are found in the bias function:
$\beta(A)=b*e^{\frac{-(A - \hat{A})^2}{2J}}$

## Chi-Square Testing

```python
def chi_squared_test_statistic(observations,expected,reps):
    O = np.zeros((N+1,N))
    for i in range(N+1):
        O[i] = np.count_nonzero(observations==i,axis=0)     
    E = expected * reps
    E[E == 0] = 1e-300
    chi_squared_hat = (((O-E)**2)/E).sum(axis=0)
    return chi_squared_hat
    
    
        
        

m = 300
b = 1
J = 1
alphas = np.full(N,0.5)
biased_indicators = bias_function(b, J, optimal_value, indicators)

expected_results = dirichlet_multinomial_probs(biased_indicators)
observations_dir = choose_dirichlet_multinomial(biased_indicators,reps=m)
observations = choose_role_model(indicators, b, J, optimal_value,alphas,eta=0.01,reps=m)

chi_square_model = chi_squared_test_statistic(observations,expected_results,m)
chi_square_dir = chi_squared_test_statistic(observations_dir,expected_results,m)
```

```python
plt.hist(np.log(chi_square_model),label='model',bins=30)
plt.hist(np.log(chi_square_dir),label='dirichlet',bins=30,alpha=0.4)
plt.xlabel(r'log($\chi^2$)')
plt.legend();
plt.show();
```

```python
import scipy

def out_limit(param,upper_bound):
    return param <= 0 or param >= upper_bound

def loss_dir(θ,A):    
    b, J, eta, alpha = θ
    if out_limit(b,1000) or out_limit(J,1000) or out_limit(eta,1) or out_limit(alpha,1):
        return np.inf
    
    beta_A = bias_function(b, J, optimal_value, A)
    alphas = np.full(len(A),alpha)
    
    K_dir = choose_dirichlet_multinomial(beta_A,reps=100).mean(axis=0)
    K_process = choose_role_model(A, b, J, optimal_value,alphas,reps=100).mean(axis=0)
    
    loss = metrics.mean_squared_error(K_process,K_dir)
    print(loss)
    return loss

θ0 = (1,1,0.01,0.5)
θ_hat = scipy.optimize.fmin(
    func=loss_dir, 
    x0=θ0,
    args=(indicators,),
    maxiter=50,
)

b_hat,J_hat,eta_hat,alpha_hat = θ_hat
print(f'b = {b_hat:.5f}\tJ = {J_hat:.5f}\teta = {eta_hat:.5f}\talpha = {alpha_hat:.5f}')
    
```

### Computing the RMSE and KL divergence of the dirichlet and GBD approximations


```python
from sklearn import metrics
from scipy.stats import dirichlet

RMSEs = {}
RMSEs_dir = {}
KLs = {}
KLs_dir = {}
alphas = np.full(N,0.5)
for err_scale in err_scales:
    rmse_matrix = []
    rmse_matrix_dir = []
    kl_matrix = []
    kl_matrix_dir = []
    for b in bs:
        j_rmse = []
        j_rmse_dir = []
        j_kl = []
        j_kl_dir = []
        for J in Js:
            biased_indicators = bias_function(b, J, optimal_value, indicators)
            expected_results_dir = choose_dirichlet_multinomial(biased_indicators,reps=100).mean(axis=0)
            model_results_mean = choose_role_model(indicators, b, J, optimal_value,alphas,reps=100).mean(axis=0)
            expected_results = biased_indicators / biased_indicators.mean()
            rmse = np.sqrt(metrics.mean_squared_error(model_results_mean, expected_results))
            rmse_dir = np.sqrt(metrics.mean_squared_error(model_results_mean, expected_results_dir))
            kl_divs_GBD = kl_divergence(expected_results,model_results_mean)
            kl_divs_dir = kl_divergence(expected_results_dir,model_results_mean)
            j_rmse.append(rmse)
            j_rmse_dir.append(rmse_dir)
            j_kl.append(kl_divs_GBD)
            j_kl_dir.append(kl_divs_dir)
        rmse_matrix.append(j_rmse)
        rmse_matrix_dir.append(j_rmse_dir)
        kl_matrix.append(j_kl)
        kl_matrix_dir.append(j_kl_dir)
    RMSEs[err_scale] = np.array(rmse_matrix)
    RMSEs_dir[err_scale] = np.array(rmse_matrix_dir)
    KLs[err_scale] = np.array(kl_matrix)
    KLs_dir[err_scale] = np.array(kl_matrix_dir)
```

```python
from matplotlib import colors
import seaborn as sns

def plot_error_measurements(maps,title,file_name,fmt="1.3f"):
    plt.rcParams.update({'font.size': 15})
    fig, axs = plt.subplots(nrows=2, ncols=2, constrained_layout=True,figsize=(16, 16))
    fig.suptitle(title, fontsize='large')

    err = err_scales[0]
    mesure_by_err = maps[err]
    ax = sns.heatmap(mesure_by_err,annot=True,xticklabels=Js,yticklabels=bs, linewidths=.35,fmt=fmt,ax=axs[0,0],norm=colors.LogNorm(vmax=mesure_by_err.max(),vmin=mesure_by_err.min()))
    ax.set_ylabel('b')
    ax.set_xlabel('J')
    ax.set_title(f'error scale: {err}')
    ax.invert_yaxis()

    err = err_scales[1]
    mesure_by_err = maps[err]
    ax = sns.heatmap(mesure_by_err,annot=True,xticklabels=Js,yticklabels=bs, linewidths=.35,fmt=fmt,ax=axs[0,1],norm=colors.LogNorm(vmax=mesure_by_err.max(),vmin=mesure_by_err.min()))
    ax.set_ylabel('b')
    ax.set_xlabel('J')
    ax.set_title(f'error scale: {err}')
    ax.invert_yaxis()

    err = err_scales[2]
    mesure_by_err = maps[err]
    ax = sns.heatmap(mesure_by_err,annot=True,xticklabels=Js,yticklabels=bs, linewidths=.35,fmt=fmt,ax=axs[1,0],norm=colors.LogNorm(vmax=mesure_by_err.max(),vmin=mesure_by_err.min()))
    ax.set_ylabel('b')
    ax.set_xlabel('J')
    ax.set_title(f'error scale: {err}')
    ax.invert_yaxis()

    err = err_scales[3]
    mesure_by_err = maps[err]
    ax = sns.heatmap(mesure_by_err,annot=True,xticklabels=Js,yticklabels=bs, linewidths=.35,fmt=fmt,ax=axs[1,1],norm=colors.LogNorm(vmax=mesure_by_err.max(),vmin=mesure_by_err.min()))
    ax.set_ylabel('b')
    ax.set_xlabel('J')
    ax.set_title(f'error scale: {err}')
    ax.invert_yaxis()

    plt.savefig(file_name)
    plt.show();
    
plot_error_measurements(RMSEs,'RMSE of general binomial distribution','heatmap_GBD.pdf')
plot_error_measurements(RMSEs_dir,'RMSE of dirichlet-multinomial distribution','heatmap_DirMul.pdf')
plot_error_measurements(KLs,'KL divergence of general binomial distribution','heatmap_KL_GBD.pdf','.3f')
plot_error_measurements(KLs_dir,'KL divergence of dirichlet-multinomial distribution','heatmap_KL_DirMul.pdf',".3f")
```

### Computing the RMSE for different indicator weights ($\alpha$)

```python
%matplotlib inline
from sklearn import metrics
import matplotlib.pyplot as plt

b = 1
J = 1
optimal_value = 1
err_scale = 0.01
biased_indicators = bias_function(b,J,optimal_value,indicators)
```

```python
n = 40
a = np.linspace(-1,1,n)
gamma = np.linspace(1,0,n)
for i in range(n):
    alphas=biased_indicators*a[i] + gamma[i]
    plt.scatter(biased_indicators,alphas)

plt.xlabel('a')
plt.ylabel(r'$\beta(A)$')
plt.show();
```

```python
rmse = []
rmse_dir = []
kl_divs_GBD = []
kl_divs_dir = []
weights = []
for i in range(n):
    indicator_weights = biased_indicators*a[i] + gamma[i]
    model_results_mean = choose_role_model(indicators, b, J, optimal_value,indicator_weights,reps=100).mean(axis=0)
    expected_results_dir = choose_dirichlet_multinomial(biased_indicators*indicator_weights,reps=100).mean(axis=0)
    expected_results = biased_indicators*indicator_weights / (indicator_weights*biased_indicators).mean()
    rmse_dir.append(np.sqrt(metrics.mean_squared_error(model_results_mean, expected_results_dir)))
    rmse.append(np.sqrt(metrics.mean_squared_error(model_results_mean, expected_results)))
    kl_divs_GBD.append(kl_divergence(expected_results,model_results_mean))
    kl_divs_dir.append(kl_divergence(expected_results_dir,model_results_mean))
```

```python
fig, (axs1,axs2) = plt.subplots(1,2,constrained_layout=True,figsize=(8,5))
fig.suptitle(r'GBD and dirichlet approximations with heterogenous $\alpha$', fontsize='large')

axs1.plot(a,rmse,label='GBD')
axs1.plot(a,rmse_dir,label='DirMul')
axs1.set(xlabel='a',ylabel='RMSE')
axs1.legend()

axs2.plot(a,kl_divs_GBD,label='GBD')
axs2.plot(a,kl_divs_dir,label='DirMul')
axs2.set(xlabel='a',ylabel='KL Divergence')
axs2.legend()


plt.savefig('hetero_alpha_measures.pdf')
```

```python
alphas = [0.001,0.01, 0.1, 0.5, 0.9, 0.99,0.999]
rmse = []
rmse_dir = []
kl_divs_GBD = []
kl_divs_dir = []
for alpha in alphas:
    indicator_weights = np.full(N,alpha)
    model_results_mean = choose_role_model(indicators, b, J, optimal_value,indicator_weights,reps=100).mean(axis=0)
    expected_results = biased_indicators / biased_indicators.mean()
    expected_results_dir = choose_dirichlet_multinomial(biased_indicators*indicator_weights,reps=100).mean(axis=0)
    rmse.append(np.sqrt(metrics.mean_squared_error(model_results_mean, expected_results)))
    rmse_dir.append(np.sqrt(metrics.mean_squared_error(model_results_mean, expected_results_dir)))
    kl_divs_GBD.append(kl_divergence(expected_results,model_results_mean))
    kl_divs_dir.append(kl_divergence(expected_results_dir,model_results_mean))

```

```python
fig, (axs1,axs2) = plt.subplots(1,2,constrained_layout=True,figsize=(8,5))
fig.suptitle(r'GBD and dirichlet approximations with homogenous $\alpha$', fontsize='large')

axs1.plot(alphas,rmse,label='GBD')
axs1.plot(alphas,rmse_dir,label='DirMul')
axs1.set(xlabel=r'$\alpha$',ylabel='RMSE')
axs1.legend()

axs2.plot(alphas,kl_divs_GBD,label='GBD')
axs2.plot(alphas,kl_divs_dir,label='DirMul')
axs2.set(xlabel=r'$\alpha$',ylabel='KL Divergence')
axs2.legend()


plt.savefig('alpha_measures.pdf')
```
